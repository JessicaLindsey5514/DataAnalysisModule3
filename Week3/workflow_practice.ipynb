### Q1. List all customers and their cities.

Return the first name, last name, and city of each customer. Sort alphabetically by last name and then by first name.

```python
# Your query here
query1 = """ SELECT first_name as first, last_name as last, city FROM customers  
"""
pd.read_sql(query1, connection)
```

### Q2. Show all products and their prices.

Display each product name along with its list price. Sort by price in descending order.

```python
# Your query here
query2 = """ SELECT product_name as product, list_price as price FROM products
"""
pd.read_sql(query2, connection)
```

### Q3. Find all customers from California.

Return first name, last name, city, and state for all customers whose state is 'CA'. Sort alphabetically by last name.

```python
# Your query here
query3 = """ SELECT first_name as first, last_name as last, city, state FROM customers WHERE state = 'CA'
"""
pd.read_sql(query3, connection)
```

### Q4. Count how many products are in each category.

Return the category name and the number of products in that category. Sort from the highest count to the lowest.

```python
# Your query here
query4 = """ SELECT count(products.product_id) as product_count, categories.category_name as category
FROM products
JOIN categories on categories.category_id = products.category_id
group by categories.category_name
order by product_count desc
"""
pd.read_sql(query4, connection)
```

### Q5. Find all orders placed in 2018.

List the order ID, order date, and customer ID for orders made during the year 2018. Sort by order date.

```python
# Your query here
query5 = """ SELECT order_id, customer_id, order_date
FROM orders
WHERE order_date LIKE '%2018%'
ORDER BY order_date ASC
"""
pd.read_sql(query5, connection)
```

### Q6. Show each order with its total number of items.

### Q7. List total revenue per store.

Revenue = quantity * list_price * (1 - discount). Join `orders`, `order_items`, and `stores`, group by store name, and return total revenue.

```python
# Your query here
query7 = """ SELECT stores.store_name, order_items.quantity * order_items.list_price  * (1 - order_items.discount) as total_revenue
FROM stores
JOIN orders ON stores.store_id = orders.store_id
JOIN order_items ON orders.order_id = order_items.order_id
GROUP BY stores.store_name
ORDER BY total_revenue DESC
"""
pd.read_sql(query7, connection)

# ordered by rev desc because it looked better 
```

### Q8. Find the top 5 customers who spent the most overall.

Join `customers`, `orders`, and `order_items`. Sum the total spending per customer and return the top five spenders.

```python
# Your query here
query8 = """ 
SELECT concat(customers.first_name, ' ', customers.last_name) as customer_name, sum(order_items.quantity * order_items.list_price * (1 - order_items.discount)) as total_spent
FROM customers
JOIN orders ON customers.customer_id = orders.customer_id
JOIN order_items ON orders.order_id = order_items.order_id
GROUP BY customer_name
ORDER BY total_spent DESC
LIMIT 5
"""
pd.read_sql(query8, connection)
```

### Q9. Show the best-selling product in each category.

Join `products`, `order_items`, and `categories`. For each category, identify the product with the highest total quantity sold.

```python
# Your query here
query9 = """ 

SELECT category_name, product_name, total_sold
FROM (
  SELECT categories.category_name, products.product_name, SUM(order_items.quantity) total_sold,
  ROW_NUMBER() OVER (PARTITION BY categories.category_name ORDER BY SUM(order_items.quantity) DESC) AS new_table
  FROM products
  JOIN order_items ON products.product_id = order_items.product_id
  JOIN categories ON products.category_id = categories.category_id
  GROUP BY categories.category_name, products.product_name) 
WHERE new_table <= 1
ORDER BY category_name, total_sold DESC;


"""
pd.read_sql(query9, connection)

# USED GOOGLE COPILOT TO HELP ON THIS ONE


 ## was messing around with this code(as a base/ this was NOT the full code I was using in the query) 
 
 # when i realized i needed a window function
 # tried, failed, and kept having issues
 # gave in and went to google

#SELECT products.product_name, categories.category_name, sum(order_items.quantity) as total_sold
#FROM products
#JOIN order_items ON products.product_id = order_items.product_id
#JOIN categories ON products.category_id = categories.category_id
#GROUP BY categories.category_name, products.product_name
#ORDER BY total_sold DESC
```

### Q10. Identify the employees (staff) who processed the most orders.

Join `staffs` and `orders`. Count the number of orders handled by each staff member and return the results sorted by highest total.

```python
# Your query here
query10 = """ 
SELECT concat(staffs.first_name, ' ', staffs.last_name) as staff_member, count(orders.staff_id) as total_orders_handled
FROM staffs
JOIN orders ON staffs.staff_id = orders.staff_id
GROUP BY staff_member
ORDER BY total_orders_handled DESC
"""
pd.read_sql(query10, connection)
```
